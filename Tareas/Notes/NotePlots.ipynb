# **Gráficos de estadistica descriptiva usando *matplotlib***

>Creamos la función para importar los datos

```python
import pandas as pd

CO2Data = pd.read_csv("../../Datos/Terminos_lagoon_TA_DIC_2023_RawData.csv")
```

> Exploramos el archivo de datos ambientales usando Pandas

```python
print(CO2Data.info())
```

> Gráfico de *histograma* simple

Se importa la libreria para gráficar

```python
import matplotlib.pyplot as plt
```

Seleccionamos los datos

```python
sal_data = CO2Data["sal_psu"]
```

Se mete la función para el gráfico

```python
plt.hist(sal_data, bins=10)
```

Mostramos el gráfico

```python
plt.show()
```

Si quiero ponele etiquetas en los ejes

```python
plt.hist(sal_data, bins=10)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probability")
plt.show()
```

- Normalizar las entradas por el número total de conteos

Gráfico de un histograma con densidad (normalizada)

```python
data = CO2Data["sal_psu"]

plt.hist(data, bins=10, density=True)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probability")
plt.show()
```

- Diagrama de caja

```python
do_data = CO2Data["do_mg_l"]

fig2, ax2 = plt.subplots()

ax2.set_title('Notched boxes')
ax2.boxplot(do_data, notch=True)

plt.ylabel('Salinity ($mg L^{-1}$)' , fontsize = 12)
plt.show()

```
