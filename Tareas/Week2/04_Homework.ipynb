## Exercise
-  ### Plot histogram of DIC and TA


> Creamos la función para importar datos

```python
import pandas as pd 
import matplotlib.pyplot as plt

CO2Data = pd.read_csv("../../Datos/Terminos_lagoon_TA_DIC_2023_RawData.csv")
```

> Seleccionamos los datos:

```python
DIC_TA_plot = CO2Data[["dic_micromol_kg", "ta_micromol_kg" ]]
```

> Se llama la función plot

```python
plt.hist(
    [DIC_TA_plot["dic_micromol_kg"], DIC_TA_plot["ta_micromol_kg"]],
    bins = 15, # clases
    alpha = 0.5, # transparecia
    label = ["dic","ta"], # nombra a cada conjunto de datos
    edgecolor = "black" # contorno en las barras
)

plt.xlabel("Valor (µmol/kg)") # etiqueta en el eje x
plt.ylabel("Frecuencia") # etiqueta en el eje y
plt.title("Histograma de dic y ta") # titulo del gráfico
plt.legend() # le pide a matplotlib que muestre las etiquetas de los datos sobre el gráfico
plt.show() # grafica todo lo que se acomulo en matplotlib 
```
