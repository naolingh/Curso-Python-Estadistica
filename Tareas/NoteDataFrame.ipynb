# **Crear un archivo CSV**

Se parte de un archivo de excel que tenga los datos que queremos trabajar y se guardan como un archivo del tipo .csv 

## Funcion para leer un archivo csv separado por comas y convertirlo en un DataFrame


```python
import pandas as pd 
def read_csv_to_dataframe(file_path):
    """
    Lee el archivo CSV y lo convierte en un DataFrame de Pandas

    Parametros:
        file_path (str): Aqui va el path del archivo CSV
    
    Regreso:
        pd.DataFrame: El DataFrame contiene los datos del archivo CSV
    """
    try:
        df = pd.read_csv(file_path)
        return df 
    except Exception as e:
        print(f"Error al leer el archivo csv: {e}")
        return None
       
```

```python
read_csv_to_dataframe("Terminos_lagoon_TA_DIC_2023_RawData.csv")
```
